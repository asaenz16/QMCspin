The following functions determine if moves are possible. There are four possible moves:
 - SEB:
 - BES:
 - SOB:
 - BOS:
 - DEU:
 - UED:
 - DOU:
 - UOD:

```python
def is_moveSEB (points , time):
    if len(points) != 4:
        return False
    cond = np.abs(points[0] - points[1]) + np.abs(points[1] - points[2]) + np.abs(points[2] - points[3])
    if cond != 0:
        return False
    cond = points[0]+time
    if cond%2 ==0:
        return True
    else:
        return False
    
def is_moveBES (points , time):
    if len(points) != 4:
        return False
    cond = np.abs(points[0]-1 - points[1]) + np.abs(points[1] - points[2]) + np.abs(points[2]+1 - points[3])
    if cond != 0:
        return False
    cond = points[0]+time
    if cond%2 ==0:
        return True
    else:
        return False
    
def is_moveSOB (points , time):
    if len(points) != 4:
        return False
    cond = np.abs(points[0] - points[1]) + np.abs(points[1] - points[2]) + np.abs(points[2] - points[3])
    if cond != 0:
        return False
    cond = points[0]+time
    if cond%2 ==1:
        return True
    else:
        return False
    
def is_moveBOS (points , time):
    if len(points) != 4:
        return False
    cond = np.abs(points[0]+1 - points[1]) + np.abs(points[1] - points[2]) + np.abs(points[2]-1 - points[3])
    if cond != 0:
        return False
    cond = points[0]+time
    if cond%2 ==1:
        return True
    else:
        return False
    
def is_moveDEU (points , time):
    if len(points) != 4:
        return False
    cond = np.abs(points[0]-1 - points[1]) + np.abs(points[1] - points[2]) + np.abs(points[2] - points[3])
    if cond != 0:
        return False
    cond = points[0]+time
    if cond%2 ==0:
        return True
    else:
        return False
    
def is_moveUED (points , time):
    if len(points) != 4:
        return False
    cond = np.abs(points[0] - points[1]) + np.abs(points[1] - points[2]) + np.abs(points[2]-1 - points[3])
    if cond != 0:
        return False
    cond = points[0]+time
    if cond%2 ==0:
        return True
    else:
        return False
    
def is_moveDOU (points , time):
    if len(points) != 4:
        return False
    cond = np.abs(points[0]+1 - points[1]) + np.abs(points[1] - points[2]) + np.abs(points[2] - points[3])
    if cond != 0:
        return False
    cond = points[0]+time
    if cond%2 ==1:
        return True
    else:
        return False
    
def is_moveUOD (points , time):
    if len(points) != 4:
        return False
    cond = np.abs(points[0] - points[1]) + np.abs(points[1] - points[2]) + np.abs(points[2]+1 - points[3])
    if cond != 0:
        return False
    cond = points[0]+time
    if cond%2 ==1:
        return True
    else:
        return False    
```

```python
def move(points, time):
    if is_moveSEB (points , time):
        return [points[0],points[1]-1,points[2]-1,points[3]]
    if is_moveBES (points , time):
        return [points[0],points[1]+1,points[2]+1,points[3]]
    if is_moveSOB (points , time):
        return [points[0],points[1]+1,points[2]+1,points[3]]
    if is_moveBOS (points , time):
        return [points[0],points[1]-1,points[2]-1,points[3]]
    if is_moveDEU (points , time):
        return [points[0],points[1]+1,points[2]+1,points[3]]
    if is_moveUED (points , time):
        return [points[0],points[1]-1,points[2]-1,points[3]]
    if is_moveDOU (points , time):
        return [points[0],points[1]-1,points[2]-1,points[3]]
    if is_moveUOD (points , time):
        return [points[0],points[1]+1,points[2]+1,points[3]]
    return points
```

```python
def random_update(wl, p):
    if random.random () > p:
        return wl
    n = len(wl)
    if n < 4:
        return wl
    temp_wl = wl.copy()
    time = random.randint(0, n-4)
    points = [wl[time], wl[time+1], wl[time+2], wl[time+3]]
    points = move(points , time)
    temp_wl[time: time+4] = points
    return (temp_wl)
```

```python
def exclusion(wl1, wl2):
    if len(wl1) != 4:
        return False
    if len(wl2) != 4:
        return False
    if len(list(set(wl1)& set(wl2))) >0:
        return False
    return True
```

```python
t = 10
n=2
steps = t**2
WL =[]
for i in range(n):
    WL.append([i]*t)
print(WL[1][0:4])
```

```python
def random_update_one(wl, p):
    if random.random () > p:
        return wl
    n = len(wl)
    if n < 4:
        return wl
    temp_wl = wl.copy()
    time = random.randint(0, n-4)
    points = [wl[time], wl[time+1], wl[time+2], wl[time+3]]
    points = move(points , time)
    temp_wl[time: time+4] = points
    return (temp_wl)

def random_update_le(wl1, wl2, p):
    if random.random () > p:
        return [wl1, wl2]
    if len(wl1) != len(wl2):
        return [wl1, wl2]
    n = len(wl1)
    if n < 4:
        return [wl1, wl2]
    temp_wl1 = wl1.copy()
    temp_wl2 = wl2.copy()
    time = random.randint(0, n-4)
    points = temp_wl1[time:time+4]
    points = move(points , time)
    if exclusion(points, temp_wl2[time:time+4]):
        temp_wl1[time: time+4] = points
    return [temp_wl1, temp_wl2]

def random_update_re(wl0, wl1, p):
    if random.random () > p:
        return [wl0, wl1]
    if len(wl0) != len(wl1):
        return [wl0, wl1]
    n = len(wl1)
    if n < 4:
        return [wl0, wl1]
    temp_wl0 = wl0.copy()
    temp_wl1 = wl1.copy()
    time = random.randint(0, n-4)
    points = temp_wl1[time:time+4]
    points = move(points , time)
    if exclusion(points, temp_wl0[time:time+4]):
        temp_wl1[time: time+4] = points
    return [temp_wl0, temp_wl1]

def random_update_b(wl0, wl1, wl2, p):
    if random.random () > p:
        return [wl0, wl1, wl2]
    if len(wl1) != len(wl2) or  len(wl1) != len(wl0) or len(wl0) != len(wl2):
        return [wl0, wl1, wl2]
    n = len(wl1)
    if n < 4:
        return [wl0, wl1, wl2]
    temp_wl0 = wl0.copy()
    temp_wl1 = wl1.copy()
    temp_wl2 = wl2.copy()
    time = random.randint(0, n-4)
    points = temp_wl1[time:time+4]
    points = move(points , time)
    if exclusion(points, temp_wl2[time:time+4]) and exclusion(temp_wl0[time:time+4], points):
        temp_wl1[time: time+4] = points
    return [temp_wl0, temp_wl1, temp_wl2]
```

```python
t = 1000
n=2
steps = 10*t**2
WL =[]
for i in range(n):
    WL.append([i]*t)
#print(WL)
for i in range(steps):
    k = random.randrange(n)
    if k==0:
        if k==n-1:
            WL[k] = random_update_one(WL[k], 0.5).copy()
        else:
            WL[k:k+2] = random_update_le(WL[k], WL[k+1], 0.5).copy()
    else:
        if k==n-1:
            WL[k-1:k+1] = random_update_re(WL[k-1], WL[k], 0.5).copy()
        else:
            WL[k-1:k+2] = random_update_b(WL[k-1], WL[k], WL[k+1], 0.5).copy()
print(len(WL))
x =range(t)            
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("A test graph")
for i in range(n):
    plt.plot(WL[i], x,label = 'id %s'%i)
#plt.legend()
#plt.xlim(-1, n)
plt.show()
```
