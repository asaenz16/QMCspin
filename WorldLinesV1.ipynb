```python
import numpy as np
import random
from matplotlib import pyplot as plt
```

The following functions determine if moves are possible. There are four possible moves:
 - SEB:
 - BES:
 - SOB:
 - BOS:
 - DEU:
 - UED:
 - DOU:
 - UOD:

```python
def is_moveSEB (points , time):
    if len(points) != 4:
        return False
    cond = np.abs(points[0] - points[1]) + np.abs(points[1] - points[2]) + np.abs(points[2] - points[3])
    if cond != 0:
        return False
    cond = points[0]+time
    if cond%2 ==0:
        return True
    else:
        return False
    
def is_moveBES (points , time):
    if len(points) != 4:
        return False
    cond = np.abs(points[0]-1 - points[1]) + np.abs(points[1] - points[2]) + np.abs(points[2]+1 - points[3])
    if cond != 0:
        return False
    cond = points[0]+time
    if cond%2 ==0:
        return True
    else:
        return False
    
def is_moveSOB (points , time):
    if len(points) != 4:
        return False
    cond = np.abs(points[0] - points[1]) + np.abs(points[1] - points[2]) + np.abs(points[2] - points[3])
    if cond != 0:
        return False
    cond = points[0]+time
    if cond%2 ==1:
        return True
    else:
        return False
    
def is_moveBOS (points , time):
    if len(points) != 4:
        return False
    cond = np.abs(points[0]+1 - points[1]) + np.abs(points[1] - points[2]) + np.abs(points[2]-1 - points[3])
    if cond != 0:
        return False
    cond = points[0]+time
    if cond%2 ==1:
        return True
    else:
        return False
    
def is_moveDEU (points , time):
    if len(points) != 4:
        return False
    cond = np.abs(points[0]-1 - points[1]) + np.abs(points[1] - points[2]) + np.abs(points[2] - points[3])
    if cond != 0:
        return False
    cond = points[0]+time
    if cond%2 ==0:
        return True
    else:
        return False
    
def is_moveUED (points , time):
    if len(points) != 4:
        return False
    cond = np.abs(points[0] - points[1]) + np.abs(points[1] - points[2]) + np.abs(points[2]-1 - points[3])
    if cond != 0:
        return False
    cond = points[0]+time
    if cond%2 ==0:
        return True
    else:
        return False
    
def is_moveDOU (points , time):
    if len(points) != 4:
        return False
    cond = np.abs(points[0]+1 - points[1]) + np.abs(points[1] - points[2]) + np.abs(points[2] - points[3])
    if cond != 0:
        return False
    cond = points[0]+time
    if cond%2 ==1:
        return True
    else:
        return False
    
def is_moveUOD (points , time):
    if len(points) != 4:
        return False
    cond = np.abs(points[0] - points[1]) + np.abs(points[1] - points[2]) + np.abs(points[2]+1 - points[3])
    if cond != 0:
        return False
    cond = points[0]+time
    if cond%2 ==1:
        return True
    else:
        return False    
```

```python
def move(points, time):
    if is_moveSEB (points , time):
        return [points[0],points[1]-1,points[2]-1,points[3]]
    if is_moveBES (points , time):
        return [points[0],points[1]+1,points[2]+1,points[3]]
    if is_moveSOB (points , time):
        return [points[0],points[1]+1,points[2]+1,points[3]]
    if is_moveBOS (points , time):
        return [points[0],points[1]-1,points[2]-1,points[3]]
    if is_moveDEU (points , time):
        return [points[0],points[1]+1,points[2]+1,points[3]]
    if is_moveUED (points , time):
        return [points[0],points[1]-1,points[2]-1,points[3]]
    if is_moveDOU (points , time):
        return [points[0],points[1]-1,points[2]-1,points[3]]
    if is_moveUOD (points , time):
        return [points[0],points[1]+1,points[2]+1,points[3]]
    return points
```

```python
def random_update(wl, p):
    if random.random () > p:
        return wl
    n = len(wl)
    if n < 4:
        return wl
    temp_wl = wl.copy()
    time = random.randint(0, n-4)
    points = [wl[time], wl[time+1], wl[time+2], wl[time+3]]
    points = move(points , time)
    temp_wl[time: time+4] = points
    return (temp_wl)
```

```python
n = 10
wl=[0]*n
print(random_update(wl, 1))
print(wl)
```
