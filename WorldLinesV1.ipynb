The following functions determine if moves are possible. There are four possible moves:
 - SEB:
 - BES:
 - SOB:
 - BOS:
 - DEU:
 - UED:
 - DOU:
 - UOD:

```python
def is_moveSEB (points , time):
    if len(points) != 4:
        return False
    cond = np.abs(points[0] - points[1]) + np.abs(points[1] - points[2]) + np.abs(points[2] - points[3])
    if cond != 0:
        return False
    cond = points[0]+time
    if cond%2 ==0:
        return True
    else:
        return False
    
def is_moveBES (points , time):
    if len(points) != 4:
        return False
    cond = np.abs(points[0]-1 - points[1]) + np.abs(points[1] - points[2]) + np.abs(points[2]+1 - points[3])
    if cond != 0:
        return False
    cond = points[0]+time
    if cond%2 ==0:
        return True
    else:
        return False
    
def is_moveSOB (points , time):
    if len(points) != 4:
        return False
    cond = np.abs(points[0] - points[1]) + np.abs(points[1] - points[2]) + np.abs(points[2] - points[3])
    if cond != 0:
        return False
    cond = points[0]+time
    if cond%2 ==1:
        return True
    else:
        return False
    
def is_moveBOS (points , time):
    if len(points) != 4:
        return False
    cond = np.abs(points[0]+1 - points[1]) + np.abs(points[1] - points[2]) + np.abs(points[2]-1 - points[3])
    if cond != 0:
        return False
    cond = points[0]+time
    if cond%2 ==1:
        return True
    else:
        return False
    
def is_moveDEU (points , time):
    if len(points) != 4:
        return False
    cond = np.abs(points[0]-1 - points[1]) + np.abs(points[1] - points[2]) + np.abs(points[2] - points[3])
    if cond != 0:
        return False
    cond = points[0]+time
    if cond%2 ==0:
        return True
    else:
        return False
    
def is_moveUED (points , time):
    if len(points) != 4:
        return False
    cond = np.abs(points[0] - points[1]) + np.abs(points[1] - points[2]) + np.abs(points[2]-1 - points[3])
    if cond != 0:
        return False
    cond = points[0]+time
    if cond%2 ==0:
        return True
    else:
        return False
    
def is_moveDOU (points , time):
    if len(points) != 4:
        return False
    cond = np.abs(points[0]+1 - points[1]) + np.abs(points[1] - points[2]) + np.abs(points[2] - points[3])
    if cond != 0:
        return False
    cond = points[0]+time
    if cond%2 ==1:
        return True
    else:
        return False
    
def is_moveUOD (points , time):
    if len(points) != 4:
        return False
    cond = np.abs(points[0] - points[1]) + np.abs(points[1] - points[2]) + np.abs(points[2]+1 - points[3])
    if cond != 0:
        return False
    cond = points[0]+time
    if cond%2 ==1:
        return True
    else:
        return False    
```

```python
def move(points, time):
    if is_moveSEB (points , time):
        return [points[0],points[1]-1,points[2]-1,points[3]]
    if is_moveBES (points , time):
        return [points[0],points[1]+1,points[2]+1,points[3]]
    if is_moveSOB (points , time):
        return [points[0],points[1]+1,points[2]+1,points[3]]
    if is_moveBOS (points , time):
        return [points[0],points[1]-1,points[2]-1,points[3]]
    if is_moveDEU (points , time):
        return [points[0],points[1]+1,points[2]+1,points[3]]
    if is_moveUED (points , time):
        return [points[0],points[1]-1,points[2]-1,points[3]]
    if is_moveDOU (points , time):
        return [points[0],points[1]-1,points[2]-1,points[3]]
    if is_moveUOD (points , time):
        return [points[0],points[1]+1,points[2]+1,points[3]]
    return points
```

```python
def random_update(wl, p):
    if random.random () > p:
        return wl
    n = len(wl)
    if n < 4:
        return wl
    temp_wl = wl.copy()
    time = random.randint(0, n-4)
    points = [wl[time], wl[time+1], wl[time+2], wl[time+3]]
    points = move(points , time)
    temp_wl[time: time+4] = points
    return (temp_wl)
```

```python
def exclusion(wl1, wl2):
    if len(wl1) != 4:
        return False
    if len(wl2) != 4:
        return False
    if len(list(set(wl1)& set(wl2))) >0:
        return False
    return True
```

```python
def random_update_one(wl, p):
    if random.random () > p:
        return wl
    n = len(wl)
    if n < 4:
        return wl
    temp_wl = wl.copy()
    time = random.randint(0, n-4)
    points = [wl[time], wl[time+1], wl[time+2], wl[time+3]]
    points = move(points , time)
    temp_wl[time: time+4] = points
    return (temp_wl)

#le = left edge
def random_update_le(wl1, wl2, p):
    if random.random () > p:
        return [wl1, wl2]
    if len(wl1) != len(wl2):
        return [wl1, wl2]
    n = len(wl1)
    if n < 4:
        return [wl1, wl2]
    temp_wl1 = wl1.copy()
    temp_wl2 = wl2.copy()
    time = random.randint(0, n-4)
    points = temp_wl1[time:time+4]
    points = move(points , time)
    if exclusion(points, temp_wl2[time:time+4]):
        temp_wl1[time: time+4] = points
    return [temp_wl1, temp_wl2]

#re = right edge
def random_update_re(wl0, wl1, p):
    if random.random () > p:
        return [wl0, wl1]
    if len(wl0) != len(wl1):
        return [wl0, wl1]
    n = len(wl1)
    if n < 4:
        return [wl0, wl1]
    temp_wl0 = wl0.copy()
    temp_wl1 = wl1.copy()
    time = random.randint(0, n-4)
    points = temp_wl1[time:time+4]
    points = move(points , time)
    if exclusion(points, temp_wl0[time:time+4]):
        temp_wl1[time: time+4] = points
    return [temp_wl0, temp_wl1]

#b = bulk
def random_update_b(wl0, wl1, wl2, p):
    if random.random () > p:
        return [wl0, wl1, wl2]
    if len(wl1) != len(wl2) or  len(wl1) != len(wl0) or len(wl0) != len(wl2):
        return [wl0, wl1, wl2]
    n = len(wl1)
    if n < 4:
        return [wl0, wl1, wl2]
    temp_wl0 = wl0.copy()
    temp_wl1 = wl1.copy()
    temp_wl2 = wl2.copy()
    time = random.randint(0, n-4)
    points = temp_wl1[time:time+4]
    points = move(points , time)
    if exclusion(points, temp_wl2[time:time+4]) and exclusion(temp_wl0[time:time+4], points):
        temp_wl1[time: time+4] = points
    return [temp_wl0, temp_wl1, temp_wl2]
```

```python
def random_update_wt_one(wl, b, n):
    n = len(wl)
    dE =0
    if n < 4:
        return [wl, dE]
    temp_wl = wl.copy()
    time = random.randint(0, n-4)
    points = [wl[time], wl[time+1], wl[time+2], wl[time+3]]
    diff = np.abs(points[1] - points[0]) + np.abs(points[2] - points[1]) + np.abs(points[3] - points[2])
    if diff == 1:
        if points[1] ==points[2]:
            points = move(points, time)
            dE =0
    if diff ==2:
        if points[0] == points[3]:
            temp_points = move(points, time)
            if temp_points == points:
                return(True)
            points = move(points, time)
            dE = (-2/n)*(np.tanh(b/n) - (1/np.tanh(b/n)))
    if diff == 0:
        p = np.tanh(b/n)*np.tanh(b/n)
        if random.random()>p:
            return [wl, dE]
        else:
            points = move(points, time)
            dE =(-2/n)*((1/np.tanh(b/n))- np.tanh(b/n))
    temp_wl[time: time+4] = points
    return([temp_wl, dE])
```

```python
def en(b):
    return (-2*sp.iv(1, b)/sp.iv(0, b))
```

```python
n=10
t = 2*n
e = 0
steps = (t**2)
avg = 100
exp_b = 0
exp =[]
wl=[0]*t
b_range = 40
step = 1
x = np.arange(0, b_range, step)
for b in x:
    exp_b=0
    for j in range(avg):
        e = -2*np.tanh(b/n)
        wl=[0]*t
        for i in range(steps):
            sol = random_update_wt_one(wl,b, n).copy()
            [wl, e] = [sol[0], e + sol[1]]
        exp_b = exp_b + e
    exp.append(exp_b/avg)
print(exp)
#plt.plot(wl, range(t))
plt.plot( x, exp)
plt.plot(x, en(x))
```
